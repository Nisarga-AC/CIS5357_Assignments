```python
# Determine the county with the highest and lowest population growth
county_highest_growth = ny_df.sort_values(ny_df['NPOPCHG2022'],ascending = False)
county_lowest_growth = ny_df.sort_values(ny_df['NPOPCHG2022'],ascending = True)
```

```python
ny_df.loc[:, 'population_growth_rate'] = (ny_df['POPESTIMATE2022'] - ny_df['POPESTIMATE2020']) / ny_df['POPESTIMATE2020']
```
