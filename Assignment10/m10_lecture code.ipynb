### CHALLENGE 1: Look at the distribution of PERC_DELIVERY in the df_summarized_orders table, then create a new column called DELIVERY_VOLUME that assigns each row to a category (none, low, medium, high) based on the value of PERC_DELIVERY. 

```python
df_summarized_orders.PERC_DELIVERY.plot(kind = "box")
```

```python
df_summarized_orders.PERC_DELIVERY.plot(kind = "hist")
```

```python
df_summarized_orders.PERC_DELIVERY.describe()
```

```python
df_summarized_orders['DELIVERY_VOLUME'] = 'NONE: 0%'
df_summarized_orders.loc[df_summarized_orders.PERC_DELIVERY > 0, 'DELIVERY_VOLUME'] = 'LOW: >0%<=10%'
df_summarized_orders.loc[df_summarized_orders.PERC_DELIVERY > 0.1, 'DELIVERY_VOLUME'] = 'MEDIUM: >10%<=15%'
df_summarized_orders.loc[df_summarized_orders.PERC_DELIVERY > 0.15, 'DELIVERY_VOLUME'] = 'HIGH: >15%'
df_summarized_orders.DELIVERY_VOLUME.value_counts()
```
