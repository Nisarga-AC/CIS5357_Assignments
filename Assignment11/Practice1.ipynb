```python
merged_df = pd.merge(orders_df, returns_df, how='left', on='Order ID')
```

```python
merged_df
```

```python
total_orders = len(merged_df['Order ID'].unique())
```

```python
total_orders
```

```python
returned_orders = merged_df[merged_df['Returned'] == 'Yes']['Order ID'].nunique()
```

```python
returned_orders
```

```python
percentage_returned = (returned_orders / total_orders) * 100
```

```python
percentage_returned
```

```python
returns_by_region = merged_df[merged_df['Returned'] == 'Yes'].groupby('Region').size() / total_orders * 100
```
