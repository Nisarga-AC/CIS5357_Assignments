A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a visual summary of key statistics and characteristics of the data, helping you understand the data's central tendency, spread, and the presence of outliers.

```python
df_superstore.Discount.plot(kind='hist',bins = 30)
plt.title("A histogram of Discount")
plt.xlabel("Discount")
```

## A histogram of Discount
The histogram shows the frequency distribution of discounts. 
The x-axis shows the discount percentage, while the y-axis shows the number of discounts in each percentage range.

The histogram shows that the most common discounts are in between 0-10% range. There are also a significant number of discounts in 20% range and there are very few discounts above 30%.

The histogram also shows that the distribution of discounts is slightly skewed to the right. This means that there are more discounts in the lower percentage ranges than in the higher percentage ranges.

```python
plt.bar(df_superstore.Region,df_superstore.Sales)
plt.title("A bar chart with Sales by Region")
plt.xlabel("Regions")
plt.ylabel("Sales")
```

## A bar chart with Sales by Region
The bar chart represents sales by region. 
The x-axis shows the regions, while the y-axis shows sales in that region.

The bar chart shows that the highest sale value of **South** region is **22638**, the highest sale value of **Central** region is **17499**, the highest sale value of **West** regions is **13999** and the highest sale value of **East** region is **11199**

```python
df_grouped = df_superstore.groupby('Region')['Sales'].sum().reset_index()
```

```python
plt.bar(df_grouped['Region'], df_grouped['Sales'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Toatl Sales')
plt.title('Sales by Region')
plt.show()
```

## A bar chart with Sales by Region
The bar chart represents sales by region. The x-axis shows the regions, while the y-axis shows the total sales in that region.

The bar chart shows that the **West** region has the highest sales, followed by the **East** region and the **Central** region. The **South** region has the lowest sales.
