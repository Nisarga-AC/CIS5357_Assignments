```python
import numpy as np

def find_outliers_tukey ( x):
    q1 = np . percentile (x , 25)
    q3 = np . percentile (x , 75)
    iqr = q3 - q1
    floor = q1 - 1.5* iqr
    ceiling = q3 + 1.5* iqr
    outlier_indices = list (x. index [( x < floor ) |( x > ceiling ) ])
    outlier_values = list (x[ outlier_indices ])

    return outlier_indices , outlier_values
```

```python
tukey_indices , tukey_values = find_outliers_tukey (heart_df['Heart Attack Risk'])
```

```python
# Calculate the proportions of diabetes, family history, and previous heart problems by Heart Attack Risk
diabetes_proportions = heart_df.groupby('Heart Attack Risk')['Diabetes'].value_counts(normalize=True)[:, 1]
family_history_proportions = heart_df.groupby('Heart Attack Risk')['Family History'].value_counts(normalize=True)[:, 1]
previous_heart_problems_proportions = heart_df.groupby('Heart Attack Risk')['Previous Heart Problems'].value_counts(normalize=True)[:, 1]

# Create a DataFrame for plotting
data = pd.DataFrame({
    'Diabetes': diabetes_proportions,
    'Family History': family_history_proportions,
    'Previous Heart Problems': previous_heart_problems_proportions
})

# Create a stacked bar chart
data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Proportion of Patients with Diabetes, Family History, and Previous Heart Problems by Heart Attack Risk')
plt.xlabel('Heart Attack Risk')
plt.ylabel('Proportion')
plt.xticks([0, 1], ['No Heart Attack', 'Heart Attack Patients'], rotation=0)
plt.legend(title='Medical History',bbox_to_anchor = (1,1))
plt.show()
```

### Findings:

#### Proportion of Patients with Diabetes, Family History, and Previous Heart Problems by Heart Attack Risk :

The provided graph illustrates the proportion of patients with diabetes, family history, and previous heart problems based on their risk of experiencing a heart attack. It demonstrates that individuals with diabetes, family history, or previous heart problems have a significantly higher likelihood of having a heart attack compared to those without these risk factors.

**Key takeaways from the graph:**

1. **Diabetes:** Patients with diabetes have the highest proportion of experiencing a heart attack among the three groups. This underscores the strong association between diabetes and cardiovascular disease, particularly heart attacks.

2. **Family history:** Individuals with a family history of heart attacks, even without any personal history of heart disease, have a higher risk of developing heart problems. This highlights the hereditary component of heart disease and the importance of early screening and preventive measures.

3. **Previous heart problems:** Individuals with a history of heart attacks have the highest risk of experiencing another heart attack. This emphasizes the importance of ongoing care and lifestyle modifications for individuals who have previously faced heart issues.

In conclusion, the graph clearly demonstrates the increased risk of heart attacks for individuals with diabetes, family history, or previous heart problems. These factors serve as crucial indicators for assessing cardiovascular risk and implementing preventive measures to minimize the risk of heart attacks.

### Findings:

#### Distribution of Triglyceride Levels by Heart Attack Risk :

The graph shows the distribution of triglyceride levels by heart attack risk. The x-axis shows triglyceride levels in milligrams per deciliter (mg/dL), and the y-axis shows the density of people with each triglyceride level. The density is represented by the height of the bars on the graph.

The graph shows that the higher the triglyceride levels, the higher the heart attack risk. The bar for people with no heart attack risk is tallest at the lowest triglyceride levels, and the bars for people with higher heart attack risk get taller as the triglyceride levels increase.

The graph also shows that people who have triglyceride levels below 200 mg/dL are considered as healthy triglyceride level. However, a group of people have triglyceride levels above 400 mg/dL. This is considered to be very high triglyceride levels, and it puts people at an increased risk of heart attack.

The graph also shows that there are some people with high triglyceride levels who do not have a heart attack, and some people with low triglyceride levels who do have a heart attack. This is because there are other factors that can also contribute to heart attack risk, such as cholesterol levels, blood pressure, and diabetes.

Overall, the graph shows that there is a strong relationship between triglyceride levels and heart attack risk. People with higher triglyceride levels have a higher risk of heart attack. It is important to keep triglyceride levels in a healthy range to reduce the risk of heart attack.
